```python
What is NumPy & Why It Is Required
Definition

NumPy (Numerical Python) is a high-performance numerical computing library used for handling large multi-dimensional arrays, performing vectorized operations, 
and enabling fast mathematical calculations.

    Why NumPy Is Required in Data Analytics
| Need                     | How NumPy Helps                                                            |
| ------------------------ | -------------------------------------------------------------------------- |
| Fast data processing     | NumPy uses optimized C-based operations (10–50x faster than Python lists). |
| Vectorized operations    | Perform calculations on entire datasets without loops.                     |
| Large-scale analytics    | Handles millions of records efficiently.                                   |
| Foundation for libraries | Libraries like Pandas, SciPy, Scikit-learn, TensorFlow are built on NumPy. |

Real Use Cases

Retail: analyzing daily/monthly/seasonal sales.

Healthcare: patient vitals statistical analysis.

Finance: stock price movement, portfolio returns.

Manufacturing: sensor data monitoring and real-time analytics.

Creating NumPy Arrays
Importing NumPy
```

```python
4.1 Lists (20 mins)

Lists are ordered, changeable, and allow duplicate values.
They are widely used to store rows of data.
```

```python
#Creating a List
numbers = [10, 20, 30, 40]
cities = ["Bareilly", "Moradabad", "Lucknow"]

#Accessing List Items (Indexing)
print(cities[0])   # Bareilly
print(cities[-1])  # Lucknow

#Changing Items
cities[1] = "Delhi"
print(cities)

#Adding & Removing Items
cities.append("Kanpur")
cities.remove("Lucknow")
print(cities)

#Useful List Functions (Very important for analytics)
marks = [78, 90, 45, 88]

print(len(marks))      # Total items
print(max(marks))      # Highest
print(min(marks))      # Lowest
print(sum(marks))      # Total score
```

```python
Mini Project: Employee Bonus Calculation

List of employee salaries:

[25000, 40000, 55000, 30000, 45000]
If salary < 30000 → bonus = 20%

If salary 30000–50000 → bonus = 10%

Else → bonus = 5%
Students must calculate final payout.
```

```python
salaries = [25000, 40000, 55000, 30000, 45000]

for s in salaries:
    if s < 30000:
        bonus = s * 0.20
    elif s <= 50000:
        bonus = s * 0.10
    else:
        bonus = s * 0.05

    print("Salary:", s, "Bonus:", bonus, "Final:", s + bonus)
```

```python
#1. Calculate Profit
def profit(cost, selling):
    return selling - cost

print(profit(500, 850))
```

```python
#5. Clean Missing Data

#Before Pandas, teach logic using simple lists.

def clean(data):
    clean_list = []
    for value in data:
        if value is not None:
            clean_list.append(value)
    return clean_list

print(clean([10, None, 20, None, 30]))
```

```python
def calculate_total(marks):
    return sum(marks)

def calculate_average(total):
    return total / 3

def get_grade(avg):
    if avg >= 90: return "A"
    elif avg >= 75: return "B"
    elif avg >= 50: return "C"
    else: return "Fail"

marks = [88, 92, 76]

total = calculate_total(marks)
avg = calculate_average(total)
grade = get_grade(avg)

print("Total:", total)
print("Average:", avg)
print("Grade:", grade)
```

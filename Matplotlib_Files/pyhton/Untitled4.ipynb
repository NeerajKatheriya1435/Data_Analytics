```python
What is NumPy?

NumPy = Numerical Python
Used for:
✅ Fast mathematical operations
✅ Arrays (better than lists)
✅ Scientific computing
✅ Basis for Pandas, ML algorithms
```

```python
A NumPy array (ndarray) is a multi-dimensional, indexed, fixed-type container for numbers.

Think of it as:

✅ A super-fast version of Python lists, designed for:

Mathematics

Data analytics

Machine learning

Scientific computing

 Why NumPy Arrays are Better than Python Lists?

                                                                             | Feature        | Python List   | NumPy Array              |
| -------------- | ------------- | ------------------------ |
| **Speed**      | Slow          | ⚡ Very fast              |
| **Memory**     | More memory   | Uses less memory         |
| **Operations** | Loop required | Direct vector operations |
| **Data Type**  | Mixed allowed | Single type (faster)     |

                                                                             
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

```python
Types of NumPy Arrays

1D Array – like a list
[10, 20, 30]

2D Array – like an Excel table
[[1, 2, 3],
 [4, 5, 6]]
```

```python
arr = np.array([10, 20, 30])

print(arr + 5)   # Broadcasting
print(arr * 2)
print(arr / 10)
```

```python
#1.5 Mathematical Functions
arr = np.array([10, 20, 30, 40])

print(arr.max())
print(arr.min())
print(arr.mean())
print(np.sum(arr))
```

```python
import matplotlib.pyplot as plt
import numpy as np


langauges=["JavaScript","Python","Java","C/C++","Dart"]
growthLevel1=[70,80,85,67,50] # 2024
growthLevel2=[54,87,45,62,88] # 2025
colors=["r","y","b","g","c"]
width=0.3
# plt.bar(langauges,growthLevel1)
# plt.bar(langauges,growthLevel2)

p1=np.arange(len(langauges))
p2=[j+width for j in p1]
# plt.bar(langauges,growthLevel,color=colors,edgecolor="b",linewidth=4,linestyle=":",alpha=0.4,width=0.4)
plt.bar(p1,growthLevel1,width,edgecolor="b",label="Language1")
plt.bar(p2,growthLevel2,width,edgecolor="g",label="Language2")

plt.legend(loc=2)
plt.xlabel("<-- Languages -->")
plt.ylabel("<-- Growth -->")
plt.title("Language Popularity Bar Garph")

# plt.xticks(p1+width/2,langauges,rotation=10)

plt.xticks(p1+width/2,langauges)
plt.show()
```

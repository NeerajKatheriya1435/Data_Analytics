```python
import numpy as np
var1=np.array([1,2,3,4])
var2=np.array([3,6,2,7])

varAdd1=var1+var2
varSub2=var1-var2
varMul3=var1*var2
varDiv4=var1/var2

print("Addition is : ",varAdd1)
print("Subtraction is : ",varSub2)
print("Multilication is : ",varMul3)
print("Divide is : ",varDiv4)
```

```python
import numpy as np
var1=np.array([1,2,3,4])

print("Vector plus: ",var1+2)
print("Vector minus: ",var1-1)
print("Vector multiply: ",var1*3)
print("Vector divide: ",var1/2)
print("Vector floor divide: ",var1//2)  # 7//3=2.33=2  9//4=2
print("Vector Modulous: ",var1%3) #
print("Vector Modulous: ",var1**2)
```

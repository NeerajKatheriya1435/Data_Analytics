```python
# Randomly shuffles elements of an array in-place (original array changes).

import numpy as np

arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)
```

```python
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6],
                 [14, 51, 16],
                 ])

np.random.shuffle(arr2) # Rows shuffle, columns stay intact
print(arr2)
```

```python
# Returns unique elements from an array (sorted by default).
# arr = np.array([1, 2, 2, 3, 4, 4, 5])
# print(np.unique(arr))


unique, counts = np.unique(arr, return_counts=True)
print(unique)
print(counts)
```

```python
# Changes the shape and size of an array
# If size increases â†’ elements repeat

arr = np.array([1, 2, 3, 4])
print(np.resize(arr, (3, 3)))
```

```python
# Converts a multi-dimensional array into 1D

arr2 = np.array([[1, 2],
                 [3, 4]])

flat = arr2.flatten()
flat[0] = 100

print(arr2)
print(flat)
```

```python
# Also converts array into 1D, but returns a view if possible

rav = arr2.ravel()
rav[0] = 200

print(arr2)
print(rav)
```

```python
# Insert in 1D Array

import numpy as np

arr = np.array([10, 20, 30, 40])
# new_arr = np.insert(arr, 2, 99)
# print(new_arr)

# Insert multiple values

arr1=np.insert(arr, 1, [5, 6])
print(arr1)
```

```python
# Insert in 2D Array (Row-wise)

arr2 = np.array([[1, 2],
                 [3, 4]
                ])

row = [9, 9]

print(np.insert(arr2, 1, row, axis=0))
```

```python
# Insert in 2D Array (Column-wise)

col = [8, 8]
print(np.insert(arr2, 1, col, axis=1))
```

```python
import pandas as pd

df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]})
df2=pd.DataFrame({"C":[10,20,30,40],"D":[12,12,13,14]})

df1.join(df2)
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]})
df2=pd.DataFrame({"C":[10,20],"D":[12,12]})

df1.join(df2)
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]},index=["a","b","c","d"])
df2=pd.DataFrame({"C":[10,20],"D":[12,12]},index=["a","b"])

# df1.join(df2)
df2.join(df1)
# df2.join(df1,how="left")
# df2.join(df1,how="right")
# df2.join(df1,how="outer")
df2.join(df1,how="inner")
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,22,33,44]},index=["a","b","c","d"])
df2=pd.DataFrame({"A":[10,20],"B":[12,12]},index=["a","b"])

# df1.join(df2,how="inner",lsuffix="_1")
df1.join(df2,how="inner",lsuffix="_1",rsuffix="_2")
```

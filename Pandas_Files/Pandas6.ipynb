### Merge DataFames

```python
import pandas as pd

var1=pd.DataFrame({"A":[1,2,3,4],"B":[2,4,7,1]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[12,14,17,11]})

# print(var1)
# print(var2)

# pd.merge(var1,var2)

# pd.merge(var1,var2,on="A")

pd.merge(var1,var2,how="inner")

# var2
```

```python
var1=pd.DataFrame({"A":[1,2,3,6],"B":[2,4,7,1]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[12,14,17,11]})

pd.merge(var1,var2,how="left")
# pd.merge(var2,var1,how="right")
# pd.merge(var2,var1,how="outer")
```

```python
var1=pd.DataFrame({"A":[1,2,3,6],"B":[2,4,7,1]})

var2=pd.DataFrame({"A":[1,2,3,5],"B":[12,14,17,11]})

# pd.merge(var1,var2,left_index=True,right_index=True)
# pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("python","java"))
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("-1","-2"))
```

```python
sr1=pd.Series([1,2,3,4,6])
sr2=pd.Series([11,12,13,14])

pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({"A":[1,2,3,6],"B":[2,4,7,1]})

df2=pd.DataFrame({"A1":[1,2,3,5],"B1":[12,14,17,11]})

pd.concat([df1,df2])
pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({"A":[1,2,3,6],"B":[2,4,7,1]})
df2=pd.DataFrame({"A":[1,2],"B":[12,14]})

# pd.concat([df1,df2],axis=1,join="inner")

# pd.concat([df1,df2],axis=1,join="outer")

pd.concat([df1,df2],axis=1,join="outer",keys=["Data1","Data2"])
```

```python
df1=pd.DataFrame({"A":[1,2,3,6]})
df2=pd.DataFrame({"A":[1,2],"B":[12,14]})

pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({"D":[1,2,3,6]})
df2=pd.DataFrame({"A":[1,2],"B":[12,14]})

pd.concat([df1,df2])
```
